### Print the extracted text

***learn***

```python
from PyPDF2 import PdfReader

reader = PdfReader("sample.pdf")
page = reader.pages[0]
print(page.extract_text())
```

***customerize***

```python
len(reader.pages)
```

```python
reader = PdfReader("sample.pdf")
content=' '
for i in range(len(reader.pages)):
    content+=reader.pages[i].extract_text()

print(content)
```

## 2. Convert the Text into Speech
- Use **pyttsx3** OR **gTTS**

### Import the library

```python
import os
from gtts import gTTS
```

### Convert the text

```python
language = 'en'
```

```python
speech = gTTS(text = content, lang = language, slow = False)
```

### Save the audio

```python
speech.save('samplepdf.wav')
```
